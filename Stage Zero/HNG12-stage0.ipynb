```python
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters
import seaborn as sns

#assigning the frequency of each unique value in Category 
a = df.Category.value_counts()
#displaying the result
print(a)

# Aggregate profits by Segment
profit_sum = df.groupby('Category')['Profit'].sum().reset_index()

#displaying the barplot of the unqiue values in sCategory according to the sum of their correponding profit column
sns.barplot(x='Profit', y='Category', data=profit_sum)
#adding a title
plt.title('Profitable categories')
```

```python
#assigning the frequency of each unique value in segment 
a = df.Segment.value_counts()
#displaying the result
print(a)

# Aggregate profits by Segment
profit_sum = df.groupby('Segment')['Profit'].sum().reset_index()

#displaying the barplot of the unqiue values in segment according to the sum of their correponding profit column
sns.barplot(x='Profit', y='Segment', data=profit_sum)
#adding a title
plt.title('Profitable Customer Segment')
```

## top-selling products

```python
df["Product"] = df.apply(lambda row: (
    f"{row['Sub-Category']} with Office Supplies" if row['Category'] == 'Office Supplies' 
    else (f"{row['Sub-Category']} with Technology" if row['Category'] == 'Technology' 
    else f"{row['Sub-Category']} with Furniture")
), axis=1)

df["Product"].value_counts()
```

```python
# Aggregate profits by Product
profit_sum = df.groupby('Product')['Profit'].sum().reset_index()

# Plot profit by product
plt.figure(figsize=(12, 8))
sns.barplot(x='Profit', y='Product', data=profit_sum)
plt.title('Total Profit by Product')
plt.xlabel('Total Profit')
plt.ylabel('Product')
plt.show()
```

```python
# Aggregate quantity by Product
qty_sum = df.groupby('Product')['Quantity'].sum().reset_index()

# Plot quantity by product
plt.figure(figsize=(12, 8))
sns.barplot(x='Quantity', y='Product', data=qty_sum)
plt.title('Total Quantity Sold by Product')
plt.xlabel('Total Quantity')
plt.ylabel('Product')
plt.show()
```

```python
top_profitable_products = profit_sum.sort_values(by="Profit", ascending=False).head(10)
print(top_profitable_products)
```

```python
famous_quantiy_products = qty_sum.sort_values(by="Quantity", ascending=False).head(10)
print(famous_quantiy_products)
```

```python
#CUSTOMER SUGMENTATION, BIGGEST BUYERS
# Aggregate sales per customer
customer_sales = df.groupby('Segment')['Sales'].sum().reset_index()

# Visualizing customer segments
plt.figure(figsize=(8, 6))
sns.barplot(x='Sales', y='Segment', data=customer_sales)
```

```python
# Aggregate sales per state
state_sales = df.groupby('State')['Sales'].sum().reset_index()

# Visualizing sales by state
plt.figure(figsize=(15, 6))
sns.barplot(x='Sales', y='State', data=state_sales.sort_values('Sales', ascending=False))
plt.title('Total Sales by State')
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.show()
```
